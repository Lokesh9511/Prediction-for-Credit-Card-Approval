# Outliers Detection

```python
# 1. Data cleaning and outlier detection

Q1 = c5['Income'].quantile(0.25)
Q3 = c5['Income'].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = (c5["Income"] < lower_bound) | (c5["Income"] > upper_bound)
```

```python
c5.loc[outliers,"Income"]=np.mean(c5["Income"])
```

```python
# 2. Data cleaning and outlier detection

Q1 = c5['Employment length'].quantile(0.25)
Q3 = c5['Employment length'].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = (c5["Employment length"] < lower_bound) | (c5["Employment length"] > upper_bound)
```

```python
c5.loc[outliers,"Employment length"]=np.mean(c5["Employment length"])
```

# Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
X=c5[['Income', 'Employment length', 'Account age', 'Has a car_Y',
       'Has a property_Y', 'Dwelling_House / apartment',
       'Dwelling_Municipal apartment', 'Dwelling_Office apartment',
       'Dwelling_Rented apartment', 'Dwelling_With parents']]

Y=c5[['Is high risk']]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
```

# Descision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(random_state=2,criterion="entropy",max_depth=10)
```

```python
clf.fit(X_train,Y_train)
```

```python
# Make predictions on the training and testing sets
ytrain_pred = clf.predict(X_train)
ytest_pred = clf.predict(X_test)
```

# Report

```python
from sklearn.metrics import accuracy_score,r2_score,classification_report
```

```python

# Display accuracy scores on the training and testing sets
print("Training Accuracy:", round(accuracy_score(Y_train, ytrain_pred) * 100))
print("Testing Accuracy:", round(accuracy_score(Y_test, ytest_pred) * 100))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Accuracy scores
training_accuracy = 98
testing_accuracy = 98

# Labels for your bars
labels = ['Training', 'Testing']

# Values corresponding to the labels: accuracy scores
values = [training_accuracy, testing_accuracy]

# Creating the bar plot
plt.figure(figsize=(8, 6)) 
sns.barplot(x=labels, y=values, palette='Blues_d')

# Adding the title and labels for clarity
plt.title('Model Accuracy Scores')
plt.xlabel('Dataset')
plt.ylabel('Accuracy (%)')

# Displaying the values on top of the bars
for i, value in enumerate(values):
    plt.text(i, value - 5, f'{value}%', ha='center', va='bottom', fontsize=12, color='white')

plt.ylim(0, 110)  # Set y-axis limits to accommodate the percentage values nicely

plt.show()
```

```python
cr=classification_report(Y_train,ytrain_pred)

print(cr)
```

```python
cr1=classification_report(Y_test,ytest_pred)

print(cr1)
```
